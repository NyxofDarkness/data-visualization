```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_data = pd.read_csv('csv/StudentsPerformance.csv')
temp_data.head(10)
```

I am here to see if test preparation makes a difference in test scores in math.
In this next plotting of the top 50 scores, you will see test preparation completed mimics the no test preparation completion, but has higher outliers.

```python

visual = sns.catplot(y='math score', x='test preparation course', kind='swarm', data=temp_data.head(50))
visual.set(xlabel='Test Preparation Done?', ylabel='Scores')
print()
```

### Insights
Since we are seeing very few completed test preperation participants, we will plot all the test takers! I anticipate it will mirror the previous plotting, but will have more upper scoring varients on the completed side. Althoguh the data is sparce, I can see a slight trend towards higher scores with the completion of test prep. 

```python
visual = sns.catplot(y='math score', x='test preparation course', kind='swarm', data=temp_data)
# visual.set_title('Male Vs. Female Scores')
visual.set(xlabel='Test Preparation Done?', ylabel='Scores')
print()
```

I expect as a reading score goes up, so does the math score.

```python
visual = sns.barplot(y='math score', x='reading score', data=temp_data.head(50))
visual.set_title('math vs. Reading Scores')
visual.set(xlabel='reading score', ylabel='math score')
print()
```

#### Insights
There seems to be a direct correlation between reading and math scores. As one goes up, so does the other.
