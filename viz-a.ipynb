```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_data = pd.read_csv('csv/scrubbed.csv')
temp_data.head()
```

```python
df = temp_data.loc[temp_data['city'] == 'seattle']
df
```

```python
x1 = df['date posted'].value_counts().head(5)
x1
```

```python
x2 = temp_data['shape'].value_counts().head(5)
x2
```

```python
plt.plot(x1, label='UFO Sightings in Seattle', color='blue')
plt.plot(x2, label='Shapes of Sightings', color='red')
plt.legend()
plt.show()
```
