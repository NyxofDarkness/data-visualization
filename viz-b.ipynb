```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_data = pd.read_csv('csv/NZairfares.csv')
temp_data.head(10)
```

```python
## I believe overnight departure times make tickets cheaper
```

```python
sns.catplot(x='Airline', y='Airfare(NZ$)', hue='Dep. time', kind='point', data=temp_data.head(30))
```

```python
### Insights: As we see here, the more we add data, the more we see it has more to do with the airline brand than the time of day! It would be irresponsible to think this is cause wholly by the airline or time though. It could also be that the two companies fly vastly different lengths, or that one caters to a more fancy cliental. 
```

```python
sns.catplot(x='Airline', y='Airfare(NZ$)', hue='Dep. time', kind='point', data=temp_data.head(100))
```

```python
data = temp_data.groupby('Duration')['Airfare(NZ$)'].sum()
data
```
